```python
from numpy import *
import operator
import seaborn as sns
sns.set(context='poster', style='dark')
sns.set_context(rc={'lines.markeredgewidth': 0.5})
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
def file_to_matrix(filename):
    file_read = open(filename)
    file_lines = file_read.readlines()
    number_of_lines = len(file_lines)
    return_mat = zeros((number_of_lines, 3))

    class_label_vector = []
    index = 0
    for line in file_lines:
        line = line.strip()
        list_from_line = line.split('\t')
        return_mat[index,:] = list_from_line[0:3]
        class_label_vector.append(int(list_from_line[-1]))
        index += 1
    return return_mat, class_label_vector
```

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    
    #distance calculation:
    diffMat = tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    
    sortedDistIndices = distances.argsort()
    classCount = {}
    
    for i in range(k): 
        voteIlabel = labels[sortedDistIndices[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    
    sortedClassCount = sorted(classCount.items(),
                              key=operator.itemgetter(1), 
                              reverse=True)
    return sortedClassCount[0][0], sortedClassCount
```

```python
def make_sets(norm_df, test_portion):
    import random as rnd

    tot_ix = range(len(norm_df))
    test_ix = sort(rnd.sample(tot_ix, int(test_portion * len(norm_df))))
    # print(len(text_ix))
    train_ix = list(set(tot_ix) ^ set(test_ix))
    # print(test_ix)
    # print(train_ix)

    test_df = norm_df.ix[test_ix]
    train_df = norm_df.ix[train_ix]
    
    return train_df, test_df

#     print(test_df)

#     print('len(test)', len(test_df))
#     print('len(train)', len(train_df))

#     print('\ntest:\n', test_df.head(20), sep='')
#     print('\ntrain:\n', train_df.head(), sep='')
```

```python
def test_model():    
    train_df, test_df = make_sets(norm_df, 0.1)
    # print(train_df.head())
    # print(test_df.head())

    train_array = zeros((len(train_df),3))
    test_array = zeros((len(test_df),3))

    cols = train_df.columns

    # print(cols[1])

    for i in range(len(cols)-1):
    #     print(i, cols[i])
        train_array[:,i] = train_df[cols[i]].values
        test_array[:,i] = test_df[cols[i]].values

    labels = train_df.liked.values

    test_df = test_df.reset_index()
    # print(test_df)

    pass_counter = 0
    fail_counter = 0

    for i in range(len(test_array)):
        result, full_stack = classify0(test_array[i,:], train_array, labels, 10)
        if result == test_df.ix[i].liked:
            pass_counter += 1
    #         print(test_array[i,:], result, 'pass')
        else:
            fail_counter += 1
    #         print(test_array[i,:], result, 'fail')

#     print('%i passes, %i fails' % (pass_counter, fail_counter))
    return pass_counter, fail_counter
    #     print(test, result)

    # test = [0.315340, 0.315201, 0.109748]

    # result, full_stack = classify0(test, train_array, labels, 10)

pass_hist = []
pass_avg = []
    
for i in range(1000):
    passes, fails = test_model()
    pass_succ = passes/(passes+fails)
    
#     print(pass_succ)
    pass_hist.append(pass_succ)
    pass_avg.append(mean(pass_hist))

plot(arange(1, len(pass_hist)+1), pass_hist, label='for trial')
plot(arange(1, len(pass_avg)+1), pass_avg, 'red', label='average')

xlabel('trial number')
ylabel('accuracy of the algorithm')
legend(loc='best')
```
